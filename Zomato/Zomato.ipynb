observation : 
1.when rating is between 4.5 - 4.9 ---> Excellent
2.when rating are between 4.0-4.4--> very good
3.when rating are between 3.5-3.9--> good
3.when rating are between 3.0-3.4--> Avg
4.when rating are between 2.5-2.5--> avg
4.when rating are between 1.0-2.4--> poor



```python
rating.head()
```

```python
plt.figure(figsize=(12,10))
sns.barplot(x='Aggregate rating',y='Rating count', data=rating)
plt.show()
```

```python
plt.figure(figsize=(12,10))
sns.barplot(x='Aggregate rating',y='Rating count',hue='Rating color', data=rating, palette=['blue','red','orange','yellow','green','darkgreen'])
plt.show()
```

obs:
1.not rated count is very high
2.maximum no. of rating is between 2.5- 3.4

```python
#count plot
sns.countplot(x='Rating color',data=rating, palette=['blue','red','orange','yellow','green','darkgreen'])
```
