```python
x = df.iloc[:,[1,2,3]]
```

```python
print(x)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_x = scaler.fit_transform(x)
```

```python
scaled_x
```

```python
# Choosing the number of cluster
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps = 0.5, min_samples = 5)

# Fit the model to your data
labels = dbscan.fit_predict(scaled_x)
```

```python
labels
```

```python
labels = dbscan.labels_
labels
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Extract unique labels, excluding noise (-1)
unique_labels = set(labels) - {-1}

plt.figure(figsize=(10,8))
# Create a list of colors for plotting
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]

# Create a scatter plot for each cluster
for label, color in zip(unique_labels, colors):
    cluster_points = scaled_x[labels == label]
    sns.scatterplot(x=cluster_points[:, 0], y=cluster_points[:, 1], color=color, label='Cluster {}'.format(label))

# Plot noise points
noise_points = scaled_x[labels == -1]
sns.scatterplot(x=noise_points[:, 0], y=noise_points[:, 1], color='k', marker='x', label='Noise')

plt.legend()
plt.title('DBSCAN Clustering')
plt.show()
```
