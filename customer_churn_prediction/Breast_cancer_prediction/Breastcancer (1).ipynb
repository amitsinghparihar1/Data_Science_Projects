```python
# feature selection
```

```python
# with the following function we can select highly correlated features
# it will remove the first feature that is correlated with anything other feature

def correlation(dataset, threshold):
    correlated_columns = set()  # Store names of correlated columns
    correlation_matrix = dataset.corr()  # Create a correlation matrix

    # Loop through the columns in the correlation matrix
    for i in range(len(correlation_matrix.columns)):
        for j in range(i):
            # Check if the absolute correlation value is greater than the threshold
            if abs(correlation_matrix.iloc[i, j]) > threshold:
                column_name = correlation_matrix.columns[i]  # Get the name of the column
                correlated_columns.add(column_name)  # Add it to the set of correlated columns

    return correlated_columns
```

```python
corr_features=correlation(df, 0.9)
corr_features
```

```python
final_df=df.drop(corr_features, axis=1)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.linear_model import LogisticRegression
import numpy as np

# Create a Logistic Regression classifier
log = LogisticRegression()

# Define a range of hyperparameters for the RandomizedSearchCV
param_dist = {
    'C': np.logspace(-3, 3, 100),  # Inverse of regularization strength
    'solver': ['liblinear', 'lbfgs'],  # Solver for optimization
    'max_iter': [100, 200, 300, 400, 500],  # Maximum number of iterations
    'random_state': [0, 42]  # Random seed for reproducibility
}

# Create a RandomizedSearchCV object
random_search = RandomizedSearchCV(
    estimator=log,  # The Logistic Regression model
    param_distributions=param_dist,  # The hyperparameter options
    n_iter=100,  # Number of parameter settings to sample
    cv=5,  # Number of cross-validation folds
    scoring='accuracy',  # Scoring metric to optimize (you can change this to your desired metric)
    n_jobs=-1,  # Number of CPU cores to use for parallel processing
    verbose=2,  # Verbosity level for output
    random_state=42  # Random seed for reproducibility
)

# Fit the randomized search model to your data
random_search.fit(x_train, y_train)

# Get the best hyperparameters found by the random search
best_params = random_search.best_params_
```
